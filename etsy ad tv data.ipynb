```python
def eastern(df):
    df = df.drop(['Title', 'Brand', 
        'Product', "Co-Op Brand", "Co-Op Product", "Broadcast Week", "Broadcast Year", "TimeZone"], axis = 1)
    df["Datetime"] = df["Date"] + " " + df["Time"]
    df["Eastern Date time"] = pd.to_datetime(df["Datetime"])
    df["zone time pacific"] = df["Eastern Date time"].dt.tz_localize("US/Eastern", ambiguous = False).dt.tz_convert("US/Pacific")
    df["HourofDay zonetimepacific"] = pd.to_datetime(df["zone time pacific"]).dt.strftime("%H")
    df["Day of week zone time pacific"] = pd.to_datetime(df["zone time pacific"]).dt.weekday + 1
#et -]
    df = df.rename(columns={"Dayparts": "Dayparts eastern", "Broadcast Month":"Broadcast month eastern", "Day Of Week" : "Day Of Week eastern",
                  "Hour Of Day" : "Hour of Day Eastern"})
    df = df.drop(['Date', 'Time', 'EQ Units', 'Datetime'], axis = 1)
    df["Eastern Date time"] = df["Eastern Date time"].map("{:.02}".format)
    return df
```
